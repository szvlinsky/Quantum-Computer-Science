```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([3.0, 1.5])

plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(-1, 5); plt.ylim(-1, 5)
plt.gca().set_aspect('equal', 'box')
plt.grid(True)
plt.title("2D vector")
plt.show()
```

## 1. Dodawanie i odejmowanie wektorów

```python
import numpy as np

a = np.array([2, 4, -1])
b = np.array([1, -3, 5])

suma = a + b
roznica = a - b

print("Wektor a:", a)
print("Wektor b:", b)
print("a + b =", suma)
print("a - b =", roznica)
```

```python
import matplotlib.pyplot as plt

plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='b')
plt.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1, color='g', label='a + b')

plt.xlim(-5, 8)
plt.ylim(-5, 8)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Dodawanie wektorów w 2D")
plt.grid(True)
plt.show()
```

## 3. Iloczyn wektorowy i skalarny


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a = np.array([1, 2, 0])
b = np.array([0, 1, 2])

axb = np.cross(a, b)
bxa = np.cross(b, a)

print("a × b =", axb)
print("b × a =", bxa)

# Wykres 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Funkcja pomocnicza do rysowania wektora
def rysuj_wektor(ax, v, kolor, etykieta):
    ax.quiver(0, 0, 0, v[0], v[1], v[2], color=kolor, label=etykieta, arrow_length_ratio=0.1)

rysuj_wektor(ax, a, 'r', 'a')
rysuj_wektor(ax, b, 'b', 'b')
rysuj_wektor(ax, axb, 'g', 'a × b')
rysuj_wektor(ax, bxa, 'orange', 'b × a')

# Ustawienia osi i etykiet
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Nieprzemienność iloczynu wektorowego")
ax.legend()
plt.show()
```

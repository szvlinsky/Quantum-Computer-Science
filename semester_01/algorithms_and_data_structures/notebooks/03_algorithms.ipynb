# 1. Suma elementów tablicy


## Pseudokod
```
int A[1..n]
int sum = 0
read(A)

for i=1 to n
    sum = sum + A[i]

write(sum)

````python
# Wczytanie tablicy A od użytkownika, dzielenie na listę według spacji, map(int, ...) konwertuje na liczby całkowite
A = list(map(int, input("Podaj elementy tablicy oddzielone spacją: ").split()))


sum = 0

for i in range(len(A)):
    sum += A[i]

print("Suma elementów tablicy:", sum)
````

# 2. Kolejne wartości potęg w tablicy


## Pseudokod
```
int x
int n
read(n, x)

real A[1...n]
A[1] = x

for int i=2 to n
    A[i] = A[i-1] * x

write(A)

````python
n = int(input("Podaj n: "))
x = int(input("Podaj x: "))

A = [0] * n
A[0] = x

for i in range(1, n):
    A[i] = A[i-1] * x

print(A)
````

# 3. Wartość wielomianu dla współczynników oraz zmiennej - algorytm Hornera

## Pseudokod
```
real A[0...n]
real x, p=0, y
read(x, A)

for int i = 0 to n
    p = p + A[i] * y
    y = y * x

write(p)

LUB

real A[0...n]
real x, p=0
read(x, A)
p = A[n]

for int i = -1 to 0
    p = p * x + A[i]

write(p)



````python
# Bez Hornera
x = float(input("Podaj x: "))
A = list(map(float, input("Podaj współczynniki A[0...n] oddzielone spacją: ").split()))

n = len(A) - 1
y = 1.0
p = 0.0

for i in range(n + 1):
    p += A[i] * y
    y *= x

print(p)
````

````python
# Algorytm Hornera
x = float(input("Podaj x: "))
A = list(map(float, input("Podaj współczynniki A[0...n] oddzielone spacją: ").split()))
n = len(A) - 1

p = A[n]
for i in range(n - 1, -1, -1):
    p = p * x + A[i]

print(p)
````

# 4. Brakująca liczba w tablicy i tablica logiczna

## Pseudokod
```
int A[1..n]
bool B[0...n]
int a
read[A]

for int i = 0 to n
    B[i] = false

for int = 1 to n
    B[A[i]] = true

int i = 0

while(B[A[i]] = true)
    i = i + 1

write(i)

````python
A = list(map(int, input("Podaj elementy tablicy A[1..n] oddzielone spacją: ").split()))
n = len(A)
B = [False] * (n + 1)

for num in A:
    B[num] = True

for i in range(n + 1):
    if not B[i]:
        print("Brakująca liczba to:", i)
        break

````

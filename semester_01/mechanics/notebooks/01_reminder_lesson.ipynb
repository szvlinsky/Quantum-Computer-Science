```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## 1. Wektor

```python
v1 = np.array([3.3, 1.5])
v2 = np.array([2.5, 6.2])

plt.figure()
plt.quiver(0, 0, v1[0], v1[1], angles='xy',color="pink", scale_units='xy', scale=0.8)
plt.quiver(0, 0, v2[0], v2[1], angles='xy',color="grey", scale_units='xy', scale=0.8)
plt.xlim(0, 16); plt.ylim(0, 10)
plt.gca().set_aspect('equal', 'box')
plt.grid(True)
plt.title("Wektor w przestrzeni dwuwymiarowej")
plt.show()
```

## 2. Dodawanie i odejmowanie wektorów

```python
suma = v1 + v2
roznica = v1 - v2
iloczyn = v1 * v2

print("Wektor v1:", v1)
print("Wektor v2:", v2)
print("v1 + v2 =", suma)
print("v1 - v2 =", roznica)
```

```python
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='b', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='brown', label='v2')
plt.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1 + v2')

plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Dodawanie wektorów w 2D")
plt.grid(True)
plt.show()
```

## 4. Iloczyn wektorowy i skalarny


```python
a = np.array([1, 2, 0])
b = np.array([0, 1, 2])

axb = np.cross(a, b)
bxa = np.cross(b, a)

# Wykres 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Funkcja pomocnicza do rysowania wektora
def vector(ax, v, kolor, etykieta):
    ax.quiver(0, 0, 0, v[0], v[1], v[2], color=kolor, label=etykieta, arrow_length_ratio=0.1)

vector(ax, a, 'r', 'a')
vector(ax, b, 'b', 'b')
vector(ax, axb, 'g', 'a × b')
vector(ax, bxa, 'orange', 'b × a')

# Ustawienia osi i etykiet
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Nieprzemienność iloczynu wektorowego")
ax.legend()
plt.show()
```
